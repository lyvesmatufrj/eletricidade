```python
import numpy as np
import pandas as pd
```

```python
def pol(z):
    return np.array([np.abs(z),(180/np.pi)*np.angle(z)])

def rec(z):
    return np.array([np.real(z),np.imag(z)])

def l_t(v,f,l_,r,l,c,g,p_c,fp,tipo=1):
    
    if tipo == 1:
        ang = np.arccos(fp)
    else: 
        ang = -np.arccos(fp)
        
    w = 2*np.pi*f

    z = r + 1j*w*l
    Z = z*l_
    y = g + 1j*w*c
    Y = y*l_

    prop = np.sqrt(z*y)
    aten = np.real(prop)
    fase = np.imag(prop)

    len_wave = (2*np.pi)/fase
    vel_wave = len_wave*f

    v_r = V/np.sqrt(3)
    i_r = (p_c/(V*fp*np.sqrt(3)))*np.exp(-1j*np.arccos(fp))

    v_s = 0.5*((v_r + i_r*Z_c)*np.exp(l_*prop) + (v_r - i_r*Z_c)*np.exp(-l_*prop))
    i_s = 0.5*((v_r/Z_c + i_r)*np.exp(l_*prop) - (v_r/Z_c - i_r)*np.exp(-l_*prop))
    
    fasores = pd.DataFrame(np.round(np.array([pol(Z),pol(Y),pol(Z_c),
                                         pol(1e3*prop),pol(v_r),
                                         pol(i_r),pol(v_s),pol(i_s)]),5),
                       columns = ['modulo','angulo'],
                       index = ['Z LT','Y LT','IMP.CARAC.','Propagação',
                               'V_R','I_R','V_S','I_S'])

    cartes = pd.DataFrame(np.round(np.array([rec(Z),rec(Y),rec(Z_c),
                                             rec(1e3*prop),rec(v_r),
                                             rec(i_r),rec(v_s),rec(i_s)]),5),
                          columns = ['real','imaginario'],
                          index = ['Z LT','Y LT','IMP.CARAC.','Propagação',
                                   'V_R','I_R','V_S','I_S'])

    wave = pd.DataFrame(np.round(np.array([len_wave,vel_wave]),5),
                        columns = ['Valor'],
                        index = ['Comprimento de onda',
                                'Velocidade de onda'])
    
    return display(fasores,cartes,wave)
```

```python
V = 500e3
f = 60
l_ = 294.84
r = 0.025 
l = 0.854e-3
c = 13.65e-9
g = 0.0375e-6
p_c = 550e6
fp = 0.85
```

```python
w = 2*np.pi*f

z = r + 1j*w*l
Z = z*l_
y = 0 + 1j*w*c
Y = y*l_
Z_c = np.sqrt(Z/Y)

prop = np.sqrt(z*y)
aten = np.real(prop)
fase = np.imag(prop)

len_wave = (2*np.pi)/fase
vel_wave = len_wave*f

v_r = V/np.sqrt(3)
i_r = (p_c/(V*fp*np.sqrt(3)))*np.exp(-1j*np.arccos(fp))

v_s = 0.5*((v_r + i_r*Z_c)*np.exp(l_*prop) + (v_r - i_r*Z_c)*np.exp(-l_*prop))
i_s = 0.5*((v_r/Z_c + i_r)*np.exp(l_*prop) - (v_r/Z_c - i_r)*np.exp(-l_*prop))
```
